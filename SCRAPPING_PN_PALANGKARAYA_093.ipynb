```python
url = 'https://putusan3.mahkamahagung.go.id/direktori/putusan/zaee7c5622e5dc1c90dd313033393438.html'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')

soup.find('table')
soup.find_all('table')

soup.find('table', class_ = 'table')
table = soup.find_all('table')

world_titles = soup.find_all('td', class_ = 'text-right')
world_table_titles = [title.text.strip() for title in world_titles]

df = pd.DataFrame(columns = world_table_titles)
column_data = soup.find_all('tr')

# Inisialisasi list kosong untuk kolom data
kolom_data = [[] for _ in range(19)]

# Mengiterasi melalui baris data
for row in column_data[1:19]:
    # Mengambil semua elemen 'td' dalam baris
    row_data = row.find_all('td')

    # Mengambil teks dari setiap 'td' dan memasukkannya ke dalam kolom yang sesuai
    for i, data in enumerate(row_data):
        if 'text-right' not in data.get('class', []):
            kolom_data[i].append(data.text.strip())

# Menghapus kolom yang kosong
kolom_data = [kolom for kolom in kolom_data if kolom]

# Cetak data dalam bentuk kolom
for kolom in kolom_data:
    print(kolom)

length = len(df)
df.loc[length] = kolom
```
